**Анализ данных о бронировании отелей**

В первом мини-проекте вам предстоит проанализировать данные о бронировании отелей. 

Hotel – тип отеля (City Hotel или Resort Hotel) 

Is canceled – бронирование было отменено (1) или нет (0); не отмененное считается успешным

Lead time – количество дней, прошедших между датой бронирования и датой прибытия  

Arrival full date – полная дата прибытия

Arrival date year – год прибытия  

Arrival date month – месяц прибытия  

Arrival date week number – номер недели прибытия

Arrival date day of month – день прибытия

Stays in weekend nights – количество выходных (суббота или воскресенье), которые гость забронировал для проживания в отеле

Stays in week nights – количество дней (с понедельника по пятницу), которые гость забронировал для проживания в отеле

Stays total nights – общее число забронированных ночей (сумма двух предыдущих колонок)

Adults – число взрослых

Children – число детей

Babies – число младенцев 

Meal – выбранный тип питания

Country – страна происхождения клиента

Reserved room type – тип зарезервированного номера

Assigned room type – тип полученного номера (может отличаться от забронированного)

Customer type – тип бронирования

Reservation status – значение последнего статуса брони: Canceled - было отменено клиентом; Check-Out - клиент
зарегистрировался, но уже покинул отель; No-Show - клиент не зарегистрировался и сообщил администрации отеля причину

Reservation status date – дата обновления статуса

```python
#название всех необработанных названий колонок в списке
names_columns = df.columns
```

```python
#функция, которая приводит названия колонок в стандартный вид и возвращает готовый словарь для переименования 
def update_name_columns(df_columns):
    update_names = []
    result_dict = {}
    for i in df_columns:
        new_i = ("_".join(i.split(" "))).lower()
        update_names.append(new_i)
    for j in range(len(df_columns)):
        result_dict[df_columns[j]] = update_names[j]
    return result_dict
```

```python
'''
или проще
def to_lower_underscore(name):
    name = name.lower.replace(' ','_')
    return name
    
df = df.rename(columns=to_lower_underscore)
'''
```

```python
#результирующий словарь
result = update_name_columns(names_columns)
result
```

```python
df = df.rename(columns=result)
```

- Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
is_canceled_countries = df \
.query('is_canceled == 0') \
.groupby('country',as_index=False) \
.agg({'is_canceled':'value_counts'}) \
.sort_values('is_canceled', ascending=False)
```

- На сколько ночей в среднем бронируют отели разных типов?

```python
#stays_total_nights, hotel
stays_total_nights_and_hotels = df \
.groupby('hotel', as_index=False) \
.agg({'stays_total_nights':'mean'}) \
.round(decimals=2)
```

```python
stays_total_nights_and_hotels
```

- Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
df['has_kids'] = df.total_kids > 0
```

```python
df.has_kids.value_counts()
```

```python
departed_users_without_kids = df \
.query('is_canceled == 1 and has_kids == 0') \
.shape[0]

departed_users_without_kids
```

```python
all_users_without_kids = df.query('has_kids == 0').shape[0]
```

```python
churn_rate_without_kids = (departed_users_without_kids / all_users_without_kids)*100
```

```python
print(f'Отношение количества ушедших пользователей (без детей) к общему количеству клиентов (без детей): {round(churn_rate_without_kids)}%')
```

```python
departed_users_with_kids = df \
.query('is_canceled == 1 and has_kids > 0') \
.shape[0]

departed_users_with_kids
```

```python
all_users_with_kids = df.query('has_kids > 0').shape[0]
churn_rate_kids = (departed_users_with_kids / all_users_with_kids)*100
print(f'Отношение количества ушедших пользователей (есть дети) к общему количеству клиентов (есть дети): {round(churn_rate_kids)}%')
```

Таким образом, churn rate выше у группы пользователей без детей.
