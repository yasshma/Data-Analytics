```python
#для данных 
scanning = False
all_data = []
```

```python
for i in open(other_path).readlines():
    
    if i.startswith('-------------'):
        scanning = False
        continue
    
    if scanning:
        if 'Meal#;' in i:
            column = i
            continue
        all_data.append([value.strip() for value in i.split(';')])
        
    if 'Non-deleted save meals' in i:
        scanning = True
        continue
```

```python
#колонки( всего 17)
final_col_names = [col.strip() for col in column.split(';') ][0:17]
#значения 
all_data
```

```python
#собираем датафрейм 
meal_df = pd.DataFrame(all_data, columns=final_col_names)
```

```python
#Немного про работу с датой и временем ( но так как файл txt с датасетом я не потеряла, то результаты выполнения кода будут скринами)
meal_df.dtypes
```

```python
#надо перевести в нормальную дату и время (типы)
meal_df['Date'] = pd.to_datetime(meal_df.Date)
meal_df = meal_df.astype({'Minutes':'int', 'Pro: float'})
```

```python
#сколько для каждой даты потрачено минут суммарно 
user_meal_df = meal_df[['Date', 'Pro', 'Minutes']]
df = user_meal_df.groupby('Date', as_index=False) \
    .agg({'Minutes': 'sum'})
```

```python
df.to_csv('df_minutes.csv', index=False)
```
