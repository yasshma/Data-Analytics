```python
def stage(worker):
    if worker[3] < 2:
        status = f'{worker[0]} {worker[1]} is junior'
    elif worker[3] >= 2 and worker[3] < 5:
        status = f'{worker[0]} {worker[1]} is middle'
    elif worker[3] >= 5:
        status = f'{worker[0]} {worker[1]} is senior'
    return status
worker = ['Olya', 'Silyutina', 350000, 3]
stage(worker)
```

```python
def del_10(numbers):
    tens = [i for i in numbers if i%10 == 0]
    return tens
values = [12, 134, 10, 47, 100, 20, 50, 160, 210]
del_10(values)
```

```python
#разбивка ( группируем по районам и применяем функцию sum к поездкам)
df.groupby('borough').aggregate({'pickups':'sum'})
```

```python
df.groupby('borough', as_index=False).aggregate({'pickups':'sum'})
#чтобы значения borough перестали быть индексами - так удобнее
```

```python
#отсортируем от max к min
df.groupby('borough', as_index=False).aggregate({'pickups':'sum'}).sort_values('pickups', ascending=False)
```

```python
# делаем более читабельно с помощью переноса
df.groupby('borough', as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False)
```

```python
df.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False)
```

```python
pickup_month_and_borough =  df.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False)
```

```python
#сохраняем в csv файл
pickup_month_and_borough.to_csv('pickup_month_and_borough_final.csv', index=False)
```

```python
df.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum', 'pcp 24': 'count'}) \
.sort_values('pickups', ascending=False)
```

```python
#пишем запросы на фильтрацию датафрейма query('запрос в качестве строки')
df \
.query('pickups > 2000') \
.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False)
```

```python
df \
.query('pickups > 2000') \
.query('hday == "Y"') \
.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False)
```

```python
#чтобы поменять название колонки по надобности после операций
df \
.query('pickups > 2000') \
.query('hday == "Y"') \
.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False) \
.rename(columns={'borough': 'raions'})
```

```python
df \
.query('pickups > 2000') \
.query('hday == "Y"') \
.groupby(['borough', 'pickup_month'], as_index=False) \
.aggregate({'pickups':'sum'}) \
.sort_values('pickups', ascending=False) \
.rename(columns={'borough': 'raions'})
```
